## Напишем простенький код для знакомства с библиотекой


```python
from fasthtml.common import *
```

```python
from fasthtml.common import *

app = FastHTML()

@app.get("/")
def page():
    return Div(H1("Привет, это мой первый сайт"), P("Не чморите пожалуйста"))

serve()
```

H1 и P здесь - то же самое что написать `<h1>text<h1>` или `<p>text<p>`


Этот простенький код создает вот такую страницу.Снимок экрана 2024-12-25 в 12.41.45.png

При запуске запустится автоматическое логирование, которое скажет вам, где именно искать сайт (изначально он локальный), но, конечно, можно использовать свой адрес при наличии.

По аналогии можно добавить, например ссылку, в HTML это делается строчкой вида `<a href=... target=...>text<\a>`, с помощью библиотеки - `Div(A(text, href=...))`. href может быть равен ссылке на страницу в интернете, либо относительному пути на нашем сайте. Например:


```python
from fasthtml.common import *

app, rt = fast_app()


@rt("/")
def page():
    return Div(H1("Привет, это мой первый сайт"), A("Другая страница", href="/another_one"))

@rt("/another_one")
def new_page():
    return Div(H1("Круто?"))

serve()
```

Или так:


```python
from fasthtml.common import *

app, rt = fast_app()


@rt("/")
def page():
    return Div(H1("Начальная страница гугла нравится мне больше"),
               A("Го туда", href="https://www.google.com", hx_target="_blank"))

serve()
```

## Усложним задачу

Того, что было описано выше, хватит, чтобы написать полноценный сайт(например, каталог). Для усложнения задачи научимся, во-первых добавлять картинки.


Сначала, чтобы программа всегда находила нужную нам картинку, сделаем статичную директорию. Создаем папку `images` в том же месте, где и исполняемый файл, помещаем туда нужную картинку и делаем так: (библиотека starlette устанавливается вместе с FastHTML)

```python
from fasthtml.common import *
from starlette.staticfiles import StaticFiles

app, rt = fast_app()
app.mount('/images', StaticFiles(directory='images'))
```

Теперь нам достаточно будет указать путь относительно исполняемого файла, а именно `"/images/name_of_picture"`.
Воспользуемся этим:


```python
from fasthtml.common import *
from starlette.staticfiles import StaticFiles

app, rt = fast_app()
app.mount('/images', StaticFiles(directory='images'))


@rt("/")
def page():
    return Div(H1("Сайт с цензурой"),
               A("Го туда", href="/another_one"))

@rt("/another_one")
def new_page():
    return Div(H1("Круто?"),
               Card(Img(src="/images/IMG_3905.PNG"), P("Цензура")))

serve()
```

## Продолжаем усложнение


Пока что сайт, который мы можем сделать, статический, пользователь может на него только смотреть и бродить по страницам. Давайте добавим действия, доступные пользователю. Например, давайте сделаем веб-приложение, где пользователь может составлять расписание дня.


Для этого вместо `Div` можно использовать другие функции.

```python
from fasthtml.common import *

app, rt = fast_app()

cases = []

@rt("/")
def page():
    add_case = Form(
        Input(type="text", name="case"),
        Button("Add"),
        method="post",
        action="/add_case"
    )
    case_list = Ul(
        *[H1(f"{cases[i]}\n") for i in range(len(cases))]
    )
    return Div(H1("My cases for the day"), add_case, case_list)


@rt("/add_case", methods=['post'])
def update_page(case: str):
    cases.append(case)
    return RedirectResponse(url="/")


serve()
```

`Form()` здесь создает поле, где можно что-то написать и кнопку, `Input()` регулирует формат входных данных. При нажатии кнопки мы остаемся на той же странице благодаря `RedirectResponse()`. Функция опять же из библиотеки `starlette`, установленной вместе с FastHTML.

## Как поменять только одну деталь сайта

Это, опять же, можно сделать огромным количеством способов, наиболее удобный из них зависит от ситуации. Например, я хочу чтобы при нажатии на одно из дел, я о нем временно забывал и пока не видел. Тогда просто в самой надписи можно добавить параметр `hx_get=`

```python
from fasthtml.common import *

app, rt = fast_app()

cases = []

@rt("/")
def page():
    add_case = Form(
        Input(type="text", name="case"),
        Button("Add"),
        method="post",
        action="/add_case"
    )
    case_list = Ul(
        *[Div(H1(f"{cases[i]}\n"), hx_get="/delete") for i in range(len(cases))] #Изменения в этой строчке
    )
    return Div(H1("My cases for the day"), add_case, case_list)


@rt("/add_case", methods=['post'])
def update_page(case: str):
    cases.append(case)
    return RedirectResponse(url="/")


@rt("/delete")
def update_page(case):
    return Div(P("Дело отложено напотом"))


serve()
```

Теперь все, кроме одной надписи не будет рисоваться заново, а она поменяется на другую. При том при обновлении сайта или добавении новой задачи скрытую задачу будет снова видно!

Снимок экрана 2024-12-25 в 19.18.25.png

А теперь я хочу, чтобы на странице всегда менялся только список дел, а остальное было отрисовано только один раз, то есть хочу поменять процесс самого добавления нового задания. Для этого нам нужно добавить немного параметров в Form.

```python
from fasthtml.common import *

app, rt = fast_app()

cases = []

@rt("/")
def page():
    add_case = Form(
        Input(type="text", name="case"),
        Button("Add"),
        method="post",
        action="/add_case",
        hx_post="/add_case",
        hx_target="#case-list",
        hx_swap="outerHTML"
    )
    case_list = Ul(
        *[Div(H1(f"{cases[i]}\n"), hx_get="/delete") for i in range(len(cases))], id="case-list"
    )
    return Div(H1("My cases for the day"), add_case, case_list)


def show_new_cases():
    return Ul(
        *[Div(H1(f"{cases[i]}\n"), hx_get="/delete") for i in range(len(cases))], id="case-list"
    )


@rt("/add_case", methods=['post'])
def update_page(case: str):
    cases.append(case)
    return show_new_cases()


@rt("/delete")
def update_page():
    return Div(P("Дело отложено напотом"))


serve()
```

`hx_target` - параметр, который отвечает за то, какая именно часть страницы будет меняться. Теперь будет меняться только `case_list` (В его определении специально есть `id=`, именно id мы указываем в `hx_target`). Теперь функция `update_page` возвращает нам только измененный список дел, и в итоге меняется только он.
